```python
# Analisis de datos countries
# Crecimiento poblacional Ecuador y Honduras
```

```python
data = pd.read_csv("countries.csv")
```

```python
data
```

```python
Ecuador = data[data.country == "Ecuador"]
```

```python
Ecuador
```

```python
Honduras = data[data.country == "Honduras"]
```

```python
Honduras
```

```python
plt.plot(Ecuador.year, Ecuador.population/10**6)
plt.plot(Honduras.year, Honduras.population/10**6)
plt.legend(["Ecuador", "Honduras"])
plt.title("Poblacion Ecuador y Honduras")
plt.xlabel("Año")
plt.ylabel("Millones de Habitantes")
plt.show()
```

```python
Ecuador.population / Ecuador.population.iloc[0]
```

```python
Ecuador.population / Ecuador.population.iloc[0]*100
```

```python
plt.plot(Ecuador.year, Ecuador.population / Ecuador.population.iloc[0]*100)
plt.plot(Honduras.year, Honduras.population / Honduras.population.iloc[0]*100)
plt.legend(["Ecuador", "Honduras"])
plt.title("Poblacion Ecuador y Honduras")
plt.xlabel("Año")
plt.ylabel("Crecimiento de la poblacion (1er año 100)")
plt.show()
```
